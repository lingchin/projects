#### 1  CEO-Request Challenge
1.1  Problem statement (summarized)
How should Olist improve its profit margin, given that it has

revenues per sellers per months
revenues per orders
reputation costs (estimated) per bad reviews
operational costs of IT system that grows with number of orders, but not linearly (scale effects)
We advise you to start with the following analysis

🕵🏻 Should Olist remove underperforming sellers from its marketplace ?

To analyse the impact of removing the worse sellers from Olist's marketplace, we will perform a what-if analysis: What would have happened if Olist had never accepted these sellers in the first place? (granted, it's hard to know in advance who is a good seller, but let's start with this and iterate later).

1️⃣ Compute, for each seller_id, and cumulated since the beginning:

the revenues it brings
the costs associated with all its bad reviews
the resulting profits (revenues - costs)
the number of orders (it will impact the overall IT costs)
2️⃣ We can then:

sort sellers by increasing profits for Olist
for each number of sellers to remove, compute the financial impact it would have made if they had never been accepted on the platform.
we may find an optimal number of sellers to remove that maximizes Olist's profit margin

#### We also noticed charging a flat 10% on sales as Olist revenue is unprofitable for 18% of Sellers; it does not cover our fixed IT costs which are BRL5 per transaction. Our recommendation is to charge a minimum sales fee of BRL5 per transaction, or 10% of the sales transaction value, whichever is the greater.

```python
#CREATE a NEW DF to INCLUDE MIN TRANSACTION COST in the REVENUE

revenue_df=sellers.copy()

#if sales x 10% per order is < IT_cost, return IT_cost x number of orders
def cover_cost(sales, n_orders):
    if sales*0.1/n_orders < IT_cost:
        return IT_cost * n_orders
    return sales*0.1

revenue_df['revenue']=revenue_df.apply(lambda x:cover_cost(x['sales'],x['n_orders']), axis=1)+revenue_df['months_on_olist']*80

revenue_df['cost']=revenue_df['cost_of_review']+revenue_df['n_orders']*IT_cost

revenue_df['profit']=revenue_df['revenue']-revenue_df['cost']
revenue_df['profit'].sum()
```

```python
revenue_df['10_percent_persales']=revenue_df.apply(lambda x:cover_cost(x['sales'],x['n_orders']), axis=1)
```

```python
plot_revenue_df=pd.DataFrame()
plot_revenue_df['profit']=[revenue_df[revenue_df['review_score']>x]['profit'].sum() for x in range_np]
plot_revenue_df['review_score']=[x for x in range_np]
```

```python
fig=plt.figure(figsize=(15,5))
#plot 1
plt.subplot(121)
plt.hist(seller_revenue['profit'], color='pink',bins=3500, edgecolor = 'HotPink',label="no transaction fee");
plt.annotate('18% of sellers\nrevenue\nbelow\nBRL5', xy = (-5, 110), 
             fontsize = 8, xytext = (-20, 150), 
             arrowprops = dict(arrowstyle="->"),
             color = 'b')
plt.ylabel("number of sellers")
plt.xlabel('profit')
plt.xlim(-50,100)
plt.ylim(0,350)
plt.grid(axis='y', linewidth=0.5)
plt.legend()

#plot2
plt.subplot(122)
plt.hist(revenue_df['profit'],color='cornflowerblue', bins=3500, edgecolor = 'DarkBlue',label='min transaction fee');
plt.annotate('18% sellers\nbreak-even\n3.34% profit\nimprovement', xy = (10, 250), 
             fontsize = 8, xytext = (20, 280), 
             arrowprops = dict(arrowstyle="->"),
             color = 'b')
plt.ylabel("number of sellers")
plt.xlabel('profit')
plt.xlim(-50,100)
plt.ylim(0,350)
plt.grid(axis='y', linewidth=0.5)
plt.legend()

fig.suptitle("profit by number of sellers")

plt.show()
```

##### By charging a min transaction fee we reduce the number of unprofitable sellers
- The improvement in profit is BRL22,307 (BRL689,915 - BRL667,608)
- 3.34% improvement in profit

## Summary of findings:
- Removing poor review score sellers slightly improved profitability as well as Olist's average review score.
- Profitability is strong within a range of seller review scores between 1.4 and 2
- Profitability is at its highest when we remove sellers with average review score below 1.5 
- Raising the benchmark cut-off beyond seller review score 2 is counter-productive, profitability is reduced.
- Removing products with review score 1 improved our profit and average review score significantly.
- Charging a flat 10% sales transaction commission fee is unprofitable for 18% of Sellers it does not cover our fixed IT costs which are BRL5 per transaction. 


## Conclusion:
- Remove sellers that score below 1.5
- Remove products that score 1
- Charge a minimum sales fee of BRL5 per transaction, or 10% of the sales transaction value, whichever sum is the greatest.
